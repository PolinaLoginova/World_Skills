```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('DataSet.xlsx')
```

```python
category = []
```

Пусть у меня будет две категории. Товары дороже 50р и дешевле 50р за единицу товара

```python
i = 0
while i < 1157:
    if df.Price[i] >= 50:
        category.append(1)
    if df.Price[i] <= 50:
        category.append(0)
    i = i + 1
```

```python
print(category)
```

Таким образом, я имею столбец 'category', представляющий собой 0 или 1 в зависимости от цены товара. Если посмотреть на исходный датасет, то видно, что разделение товаров дороже и дешевле 50р 50%/50% (точнее, около 53%/47%. Для обучения нейронной сети я разделю данные в процентном отношении 70/30,так, чтобы внутри каждой части (обущающая и тестовая выборка) сохранялось соотношение распределения категорий.
Количество товаров с ценой больше 50р определяю, вручную отсортировав таблицу. Затем собираю новые таблицы с обучающей выборкой и с тестовой выборкой.
